Q1.Write a program in Python to Implement multithreading. Create two threads first thread will display even no’s from 1to 50 and Second trade will display odd no from 1 to 50.

```python
import threading
# Function to display even numbers from 1 to 50
def print_even():
    for i in range(2, 51, 2):
        print(i)
# Function to display odd numbers from 1 to 50
def print_odd():
    for i in range(1, 51, 2):
        print(i)
# Create threads
t1 = threading.Thread(target=print_even)
t2 = threading.Thread(target=print_odd)
# Start threads
t1.start()
t2.start()
# Wait for threads to finish
t1.join()
t2.join()
# Done
print("All threads have completed.")
```

Q2.Write a program in python to implement multithreading to create 3 threads 1 to display digits 1 to 26 ,1 to display a to z and 1 to display A to Z.

```python
import threading
# Function to display squares of numbers from 1 to 20
def print_squares():
    for i in range(1, 21):
        print(f"{i} squared is {i**2}")
# Function to display cubes of numbers from 1 to 20
def print_cubes():
    for i in range(1, 21):
        print(f"{i} cubed is {i**3}")
# Create threads
t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)
# Start threads
t1.start()
t2.start()
# Wait for threads to finish
t1.join()
t2.join()
# Done
print("All threads have completed.")
```

Q3.Write a program in Python to Implement multithreading. Create two threads first thread will display square of no’s from 1to 20 and Second trade will display cube of no from 1 to 20.

```python
import threading
# Function to display digits 1 to 26
def print_digits():
    for i in range(1, 27):
        print(i)
# Function to display lowercase letters a to z
def print_lowercase():
    for i in range(ord('a'), ord('z')+1):
        print(chr(i))
# Function to display uppercase letters A to Z
def print_uppercase():
    for i in range(ord('A'), ord('Z')+1):
        print(chr(i))
# Create threads
t1 = threading.Thread(target=print_digits)
t2 = threading.Thread(target=print_lowercase)
t3 = threading.Thread(target=print_uppercase)
# Start threads
t1.start()
t2.start()
t3.start()
# Wait for threads to finish
t1.join()
t2.join()
t3.join()
# Done
print("All threads have completed.")
```
